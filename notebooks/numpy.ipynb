```python
#### EXO1 ####

#Crée un tableau NumPy contenant les nombres de 1 à 10.
T=np.arange(1,11,1)
print(T)
#Multiplie tous les éléments par 3 et affiche le résultat.
print(T*3)
#Calcule la somme, la moyenne et l’écart-type du tableau
print("somme",T.sum())
print("moyenne",T.mean())
print("ecart-type",T.std())
#Sélectionne uniquement les nombres supérieurs à 5 et affiche-les.
print(T[T>5])
```

```python
#### EXO2 ####

#Crée un tableau 2x3 contenant les nombres de 1 à 6.
T_2D=np.arange(1,7,1).reshape(2,3)
print(T_2D)
#Affiche sa forme (shape) et sa taille (size).
print("shape:",T_2D.shape," size:",T_2D.size)
#Calcule la somme des lignes et la somme des colonnes.
print("somme des lignes:",T_2D.sum(axis=1),"somme des colonnes:",T_2D.sum(axis=0))
#Remplace tous les éléments supérieurs à 4 par 0.
T_2D[T_2D>4]=0
print(T_2D)
```

```python
#### EXO3 ####

import pandas as pd
#Crée une Series avec les notes : [12, 15, 18, 10, 14] et les étiquettes ["Math", "Physique","Chimie", "Info", "Bio"].
N=[12, 15, 18, 10, 14]
S=pd.Series(N,index=["Math", "Physique","Chimie", "Info", "Bio"])
print(S)

#Affiche la note de "Chimie".
print(S["Chimie"])
#Calcule la moyenne des notes
print(S.mean())
#Sélectionne les matières dont la note est supérieure ou égale à 15.
print(S[S>15])
print(S[S.values>15])
```

```python
#### EXO4 ####

#Crée un DataFrame avec les colonnes :....
## a partir d'un dictionnaire de listes
data={'Nom':['Ali','Sara','Youssef'],
      'Age':[23,21,22],
      'Ville':['Tunis','Sfax','Carthage']}
df1=pd.DataFrame(data)
print(df1)

## a partir d'un dictionnaire de Series
Serie={'Nom':pd.Series(['Ali','Sara','Youssef']),
       'Age':pd.Series([23,21,22]),
       'Ville':pd.Series(['Tunis','Sfax','Carthage'])}
df=pd.DataFrame(Serie)
print(df)

## a partir des ndarray de numpy
arr=np.arange(12).reshape(4,3)
df=pd.DataFrame(arr,columns=['Nom','Age','Ville'],index=['L1','L2','L3','L4'])
print(df)

#Affiche uniquement la colonne Âge.
print("la colonne Age \n",df1['Age'])

#Sélectionne les lignes où l’âge est supérieur à 23.
print("les cols avec age>23\n",df1[df1['Age']>21])

#Ajoute une nouvelle colonne "Score" avec les valeurs [85, 90, 95].
df1['Score']=[85,90,95]
print(df1)

#Calcule l’âge moyen des personnes dans le DataFrame.
print("age moyen:",df1['Age'].mean())

#Ajoute une nouvelle colonne Âge+5 qui ajoute 5 à chaque âge
df1['Age+5']=df1['Age']+5
print(df1)
```
